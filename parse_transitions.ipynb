```python
from pandas import DataFrame 
array = concat.values.tolist()
```

```python
#give position in line, parse line to get station name
def parse_st(lines, i):
    station = lines[i].split('|')[0]
    return station

#give position in line, parse line to get value
def parse_value(lines, i):
    value = lines[i].split('|')[1]
    if value == 'nan':
        value = 0
    return value
```

```python
transitions = pd.DataFrame() #creates a new dataframe that's empty


for row in array:
    part_id, response = row[0],row[-1]

#    print(station_id, response)
    #remove 0s to get all lines the part moved through
    lines = list(filter(('0').__ne__, row[1:-1]))

          
    #find start and end station names
    try:
        start_line = split(lines,0)
        end_line = split(lines,-1)
    except IndexError: #line does not have any stations 
        start_line = None
        end_line = None
        
    parsed = []
    #parse each transition into its own row
    for i in range(0,len(lines)-1):
        parsed.append([
            parse_st(lines,i),         #enter station
            parse_st(lines,i+1),       #exit station
            start_line,                #line start station
            end_line,                  #line end station
            parse_value(lines,i),      #value
            part_id,                #station id
            response])                 #response

    #save parsed data to dataframe
    parsed = DataFrame.from_records(
        parsed,columns = ['enter','exit','line_start','line_end','value','part_id','response'])

    #append to master dataframe
    transitions = transitions.append(parsed, ignore_index = True) 
    
```
